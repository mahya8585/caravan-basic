## 1. 基本のデータ型

Pythonでは、値に『型』があります。まずはよく使う3つを見ていきます。

- `int`: 整数（例: `1`, `-10`, `0`）
- `float`: 小数（例: `3.14`, `-0.5`）
- `str`: 文字列（例: `'hello'`, `'123'`）

`int` と `float` の違いは、「小数点を持てるかどうか」です。
たとえば `3` は `int`、`3.0` は `float` です。

計算では `int` と `float` を混ぜると、結果は `float` になることが多いです（例: `1 + 2.0` は `3.0`）。
また、`/` は結果が `float` になり、`//` は整数部分だけを取り出します（例: `7 / 2 = 3.5`, `7 // 2 = 3`）。

`str` は見た目が数字でも文字として扱われます。
たとえば `'123'` は `int` ではなく `str` なので、数値計算の前に `int('123')` のような型変換が必要です。

型が分からないときは `type(変数名)` で確認できます。

```python
# age/height/nameの定義をする


print(age, type(age))
print(height, type(height))
print(name, type(name))
```

## 5. mutable と immutable

ここはとても大事なポイントです。
同じ「コレクション」でも、**作ったあとに中身を変えられるかどうか** が違います。

### immutable とは
作成したあとに中身を変更できないデータです。
例: `int`, `float`, `str`, `tuple`

### mutable とは
作成したあとに中身を変更できるデータです。
例: `list`, `dict`, `set`

特に `list` と `tuple` は見た目が似ていますが、性質が異なります。
- `list` は mutable なので、`append()` などで中身を変更できます。
- `tuple` は immutable なので、要素の書き換えはできません（エラーになります）。

最初のうちは、**あとから変更する予定があるなら list、固定データなら tuple** と覚えると使い分けやすいです。

```python
# list は mutable（変更できる）
print('変更前の list:', foods)
# ここに要素追加コードを書く

print('変更後の list:', foods)

# tuple は immutable（変更できない）
print('tuple1番目:', regions[0])
# tupleに要素追加するコードを書く

# ここで TypeError になる


# str も immutable（再代入で新しい文字列を作る）
university_name = '金沢'
print('university_nameのID:', id(university_name))
university_name = university_name + '大学'
print('university_nameのID（再代入後）:', id(university_name))

university_name[0] = 'K'   # TypeError: 'str' object does not support item assignment
```

## 8. import と from

Pythonでは、今自分が書いているコードやファイル以外の処理を利用したい場合、 `import` と宣言したのち、使いたい機能を記述します。

まずは『便利な道具を使えるようにする合図』くらいのイメージで大丈夫です。
たとえば次のようなものを読み込めます。

- Python本体に最初から付属している機能を使う
- 誰かが作った便利な外部パッケージを使う
- 自分が別ファイルに書いた処理を使う

`import` は、**すでにある機能を読み込んで再利用する仕組み**です。

```python
# 計算用メソッドを使えるようにする
# 乱数生成とそれにまつわる処理を使えるようにする

# 16の平方根を計算する
# 1から10までのランダムな整数を生成する
```



`as` をつけると、読み込んだ名前に別名（エイリアス）をつけられます。
たとえば `import random as rnd` と書くと、`random` の代わりに `rnd` で呼び出せます。
長い名前を短くしたいときや、名前の衝突を避けたいときに便利です。

```python
import math as m      # as で別名をつける
import random as rnd  # random を rnd という短い名前で使う

# 書き方は異なるが、同じ結果が得られる
```

### from文

`import` はメソッド全体を読み込む書き方です。例: `import random`

`from` 文は、メソッドの中から必要な機能だけを取り込む書き方です。
メソッド全体を取り込むと、場合によってはとても膨大なコードを読み込ませることとなり、処理速度に影響することがあります。これを回避するためにfrom文を使って必要な機能だけを読み込むことで、余計なコードを読む必要がなくなります。
